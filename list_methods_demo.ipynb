1. append()

```python
l_append = [1, 2, 3]
l_append.append(4)
print(f"After append(4): {l_append}")
```

2. clear()

```python
l_clear = ['a', 'b', 'c']
l_clear.clear()
print(f"After clear(): {l_clear}")
```

3. copy()

```python
original = [10, 20, 30]
l_copy = original.copy()
l_copy.append(40)
print(f"Original: {original}")
print(f"Copy: {l_copy}")
```

4. count()

```python
l_count = [1, 2, 3, 1, 1, 4]
occurences = l_count.count(1)
print(f"Number of 1s: {occurences}")
```

5. extend()

```python
l_main = [1, 2, 3]
l_extra = [4, 5, 6]
l_main.extend(l_extra)
print(f"After extend: {l_main}")
```

6. index()

```python
l_index = ['apple', 'banana', 'cherry', 'banana']
pos = l_index.index('banana')
print(f"First index of 'banana': {pos}")
```

7. insert()

```python
l_insert = [1, 2, 4, 5]
l_insert.insert(2, 3) # (index, value)
print(f"After insert at index 2: {l_insert}")
```

8. pop()

```python
l_pop = ['x', 'y', 'z']
removed = l_pop.pop(1)
print(f"Removed item: {removed}")
print(f"Remaining list: {l_pop}")
```

9. remove()

```python
l_remove = [10, 20, 30, 20]
l_remove.remove(20)
print(f"After removing first 20: {l_remove}")
```

10. reverse()

```python
l_reverse = [1, 2, 3, 4]
l_reverse.reverse()
print(f"Reversed list: {l_reverse}")
```

11. sort()

```python
l_sort = [5, 2, 9, 1]
l_sort.sort()
print(f"Sorted list: {l_sort}")
```

### Implementation of the 11 Methods

1. append() & extend()

```python
# append adds the element as a single item
l_methods = [1, 2, 3]
l_methods.append([4, 5]) 
print(f"After append: {l_methods}") # [1, 2, 3, [4, 5]]

# extend unpacks the iterable
l_methods = [1, 2, 3]
l_methods.extend([4, 5])
print(f"After extend: {l_methods}") # [1, 2, 3, 4, 5]
```

2. insert() & index()

```python
fruits = ['apple', 'cherry']
fruits.insert(1, 'banana') # Inserts 'banana' at index 1
print(f"List: {fruits}")
print(f"Index of 'cherry': {fruits.index('cherry')}") # Returns 2
```

3. count() & copy()

```python
vals = [1, 2, 2, 3, 2]
print(f"Count of 2: {vals.count(2)}") # Returns 3

# Creating a shallow copy
vals_copy = vals.copy()
# vals_copy.clear()
print(f"Original: {vals}, Copy: {vals_copy}")
```

4. pop() & remove()

```python
items = ['PC', 'Laptop', 'Tablet', 'Laptop']
removed_item = items.pop(0) # Removes by index
print(f"Popped: {removed_item}")

items.remove('Laptop') # Removes first occurrence of value
print(f"After remove: {items}")
```

5. reverse() & sort()

```python
nums = [5, 2, 9, 1]
nums.reverse()
print(f"Reversed: {nums}")

nums.sort()
print(f"Sorted: {nums}")
```

6. clear()

```python
final_list = [1, 2, 3]
final_list.clear()
print(f"Cleared list: {final_list}") # []
```
